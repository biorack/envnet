```python
from pyteomics import mgf
import numpy as np
def read_mgf(mgf_path: str) -> pd.DataFrame:
    msms_df = []
    with mgf.MGF(mgf_path) as reader:
        for spectrum in reader:
            d = spectrum['params']
            d['spectrum'] = np.array([spectrum['m/z array'],
                                      spectrum['intensity array']])
            if 'precursor_mz' not in d:
                d['precursor_mz'] = d['pepmass'][0]
            else:
                d['precursor_mz'] = float(d['precursor_mz'])
            msms_df.append(d)
    msms_df = pd.DataFrame(msms_df)
    return msms_df

mgf_file = '../data/nl_spectra.mgf'
scn_df = read_mgf(mgf_file)
scn_df['num_peaks'] = scn_df['spectrum'].apply(lambda x: len(x[0]))
scn_df
```

```python
# use_peaks=1#{"peaks": "171.11880493164062 10203.6787109375 \n175.11251831054688 10466.7529296875 \n189.1288299560547 103202.8515625 \n191.1070556640625 38107.1875 \n205.12326049804688 41915.21875 \n207.1393585205078 169112.25 \n215.10752868652344 18078.001953125 \n219.10231018066406 14840.9599609375 \n223.13426208496094 32305.900390625 \n233.11842346191406 156739.671875 \n251.12887573242188 287390.71875 \n267.12353515625 38263.73828125 \n277.1080017089844 133010.953125 \n295.08154296875 94697.390625 \n295.11834716796875 640389.6875 \n311.113525390625 13087.2177734375 \n321.09783935546875 93350.0625 "}'
import pandas as pd
import requests
import time
def fastsearch(scans_id):
    url = "https://fasst.gnps2.org/search"
    usi = "mzspec:GNPS2:TASK-5fa279ba506643e0bee5009b631abfcb-input_spectra/nl_spectra.mgf:scan:%s"%scans_id# 37904, 28971
    # usi = "mzspec:GNPS:GNPS-LIBRARY:accession:CCMSLIB00000001547"
    database = "gnpsdata_index_11_25_23"
    analog = False
    precursor_mz_tol = 0.3
    fragment_mz_tol = 0.01
    min_cos = 0.7
    params = {
        "usi": usi, 
        "charge":1,
        "library": database, 
        "analog": "Yes" if analog else "No",
        "pm_tolerance": precursor_mz_tol,
        "fragment_tolerance": fragment_mz_tol,
        "cosine_threshold": min_cos 
    }
    max_attempts = 5
    sleep_time = 5  # time to wait before retrying, in seconds

    for attempt in range(max_attempts):
        try:
            response = requests.get(url, params=params, timeout=500)
            response.raise_for_status()  # raise exception if invalid response
            # if the request was successful, exit the loop
            response = response.json()
            response = pd.DataFrame(response['results'])
            return response
        except requests.exceptions.RequestException as e:
            print(f"Request failed: {e}")
            if attempt < max_attempts - 1:  # if this wasn't the last attempt
                print(f"Waiting {sleep_time} seconds before retrying.")
                time.sleep(sleep_time)
            else:  # if this was the last attempt
                print("All attempts failed.")
                return None
                



out = {}
counter = 0
for i,row in scn_df.iterrows():
    if row['scans'] in out:
        continue
    r = fastsearch(row['scans'])
    r['original_index'] = row['scans']
    print(counter,r.shape)
    counter+=1
    out[row['scans']] = r
    


out = pd.concat(out)
# .groupby('Dataset').count()
```
